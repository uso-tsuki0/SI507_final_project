```python
import requests as req
import numpy as np
from bs4 import BeautifulSoup

# Get the html from the website
url = 'https://www.billboard.com/charts/hot-100/'
response = req.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
songs = soup.find_all('div', class_='o-chart-results-list-row-container')
board = []
for song in songs:
    rank = song.find('span', class_="c-label").text.strip()
    info = song.find('li', class_='lrv-u-width-100p')
    title = info.find('h3').text.strip()
    artist = info.find('span').text.strip()
    history = [item.text.strip() for item in info.find_all('span', class_='c-label')[0:4]]
    board.append([rank, title, artist]+history)
board
```
